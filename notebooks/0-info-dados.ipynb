## Funções para Analise Estatística do Dados Brutos

```python
def analisar_colunas(df: pd.DataFrame) -> pd.DataFrame:
    """Cria relatório completo de análise de colunas"""
    analise = []
    
    for coluna in df.columns:
        info = {
            'Coluna': coluna,
            'Tipo': str(df[coluna].dtype),
            'Total_Registros': len(df),
            'Valores_Unicos': df[coluna].nunique(),
            'Valores_Nulos': df[coluna].isnull().sum(),
            'Perc_Nulos': f'{(df[coluna].isnull().sum() / len(df) * 100):.1f}%',
            'Perc_Unicidade': f'{(df[coluna].nunique() / len(df) * 100):.1f}%',
            'Valor_Mais_Frequente': df[coluna].mode()[0] if not df[coluna].mode().empty else None,
            'Frequencia_Maior': df[coluna].value_counts().iloc[0] if len(df[coluna].value_counts()) > 0 else 0
        }
        analise.append(info)
    
    return pd.DataFrame(analise)


def salvar_relatorio_completo(
    df: pd.DataFrame,
    arquivo_md: Path,
    arquivo_csv: Optional[Path] = None,
    titulo: Optional[str] = None,
    arquivo_origem: Optional[str] = None
) -> Tuple[Path, Path]:
    """
    Salva análise de DataFrame em Markdown e CSV

    Args:
        df: DataFrame a analisar
        arquivo_md: Caminho do arquivo .md de saída
        arquivo_csv: Caminho do arquivo .csv de saída (opcional)
        titulo: Título do relatório (opcional)
        arquivo_origem: Nome do arquivo CSV original (opcional)

    Returns:
        Tupla com (caminho_md, caminho_csv)
    """
    # Gerar análise
    relatorio = analisar_colunas(df)

    # Se não especificou arquivo CSV, usar mesmo nome do MD
    if arquivo_csv is None:
        arquivo_csv = arquivo_md.with_suffix('.csv')

    # SALVAR CSV
    relatorio.to_csv(arquivo_csv, sep=';', index=False, encoding='utf-8-sig')
    print(f'Estatísticas salvas em CSV: {arquivo_csv}')


    # SALVAR MARKDOWN
    with open(arquivo_md, 'w', encoding='utf-8') as f:
        # Cabeçalho
        f.write(f'# {titulo or "Relatório de Análise de Dados"}\n\n')
        f.write('---\n\n')

        # Metadados
        f.write('## 📋 Informações Gerais\n\n')
        if arquivo_origem:
            f.write(f'- **Arquivo:** `{arquivo_origem}`\n')
        f.write(f'- **Data:** {datetime.now().strftime("%d/%m/%Y %H:%M:%S")}\n')
        f.write(f'- **Registros:** {len(df):,}\n')
        f.write(f'- **Colunas:** {len(df.columns)}\n\n')

        # Arquivos gerados
        f.write('## Arquivos Gerados\n\n')
        f.write(f'- **Relatório Markdown:** `{arquivo_md.name}`\n')
        f.write(f'- **Estatísticas CSV:** `{arquivo_csv.name}`\n\n')

        # Dicionário de Dados
        f.write('## Dicionário de Dados\n\n')
        f.write('> **Nota:** Preencher a coluna "Descrição do tipo do dado" conforme necessário.\n\n')

        # Cabeçalho da tabela
        f.write('| Coluna | Tipo | Descrição do tipo do dado |\n')
        f.write('|--------|------|---------------------------|\n')

        # Linhas da tabela
        for coluna in df.columns:
            tipo = str(df[coluna].dtype)
            f.write(f'| {coluna} | {tipo} |  |\n')

        f.write('\n')

        # Tabela principal de análise
        f.write('## 📊 Análise por Coluna\n\n')
        f.write(f'> Os dados detalhados estão disponíveis no arquivo [`{arquivo_csv.name}`]({arquivo_csv.name})\n\n')
        f.write(relatorio.to_markdown(index=False))
        f.write('\n\n')

        # Rodapé
        f.write('---\n')
        f.write('*Relatório gerado automaticamente*\n')

    print(f'Relatório Markdown salvo: {arquivo_md}')

    return arquivo_md, arquivo_csv
```

## Gera relatórios Estatístico dos Dados

```python
# Gera relatórios estatísticos
md_file, csv_file = salvar_relatorio_completo(
    df=df,
    arquivo_md=Path(file_statistical_md),
    arquivo_csv=Path(file_statistical_csv),
    titulo='Análise de Dados - Call Center',
    arquivo_origem=dados_csv
)
```
